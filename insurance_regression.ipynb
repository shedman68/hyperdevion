```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

our_data = pd.read_csv("insurance.csv")
our_data.head()
```

```python
x = our_data.iloc[:,:1].values
y = our_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Fit the model
insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

```python
# Plot the data and model
y_pred = insurance_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
